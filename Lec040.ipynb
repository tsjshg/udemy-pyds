```python
import numpy as np
from pandas import DataFrame
```

```python
# 疑似乱数列の種を与えます。

np.random.seed(12345)
dframe = DataFrame(np.random.randn(1000, 4))
```

```python
dframe.head()
```

```python
dframe.tail()
```

```python
# describe()メソッドは便利です。

dframe.describe()
```

```python
# 最初の列
# dframe[0]　としても同じです

col = dframe.iloc[:, 0]
```

```python
col.head()
```

```python
# 3より大きい要素を取り出します。

col[np.abs(col) > 3]
```

```python
np.abs(dframe) > 3
```

```python
#  axis=1としても同じです。

(np.abs(dframe) > 3).any(axis="columns")
```

```python
# どこかが3より大きい行が取り出せます。

dframe[(np.abs(dframe) > 3).any(axis="columns")]
```

```python
np.sign(dframe)
```

```python
# 絶対値の上限が3になるようにします。

dframe[np.abs(dframe) > 3] = np.sign(dframe)*3
```

```python
dframe.describe()
```

```python
# 同じデータが作られます。
np.random.seed(12345)
dframe = DataFrame(np.random.randn(1000, 4))

# 実はメソッド1つでできます。
dframe.clip(-3, 3).describe()
```
