```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Seabornにサンプルデータがあります。
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# lmplotを使った描画
sns.lmplot(tips, x="total_bill", y="tip")
```

```python
# グラフごとにパラメータを変えられます。
sns.lmplot(tips, x="total_bill", y="tip",
           scatter_kws={'marker':'o', 'color':'indianred'},
           line_kws={'linewidth':1, 'color':'blue'})
```

```python
# 4次関数で回帰曲線をひくこともできます。
sns.lmplot(tips, x="total_bill", y="tip",
           order=4,
           scatter_kws={"marker": "o", "color": "indianred"},
           line_kws={"linewidth": 1, "color": "blue"})
```

```python
# 単なるプロットもできます。
sns.lmplot(tips, x="total_bill", y="tip",
           fit_reg=False)
```

```python
# 離散的な値でもlmplot()は使えます。

# チップの割合を計算します。
tips["tip_pect"]=100*(tips['tip']/tips['total_bill'])
tips.head()
```

```python
sns.lmplot(tips, x="size", y="tip_pect")
```

```python
sns.lmplot(tips, x="size", y="tip_pect",
           x_jitter=0.2)
```

```python
# 特定の推定量を描画することもできます。
sns.lmplot(tips, x="size", y="tip_pect",
           x_estimator=np.mean)
```

```python
# hueが便利です。
sns.lmplot(tips, x="total_bill", y="tip_pect",
           hue="sex", markers=["x","o"])
```

```python
sns.lmplot(tips, x="total_bill", y="tip_pect",
           hue="day")
```

```python
# LOESSの説明（英語）です。
url = 'http://en.wikipedia.org/wiki/Local_regression'

sns.lmplot(tips, x="total_bill", y="tip_pect",
           lowess=True,
           line_kws={"color": 'black'})
```

```python
# lmplot() は実は、もっと低レベルな関数regplotを使っています。
sns.regplot(tips, x="total_bill", y="tip_pect")
```

```python
# 描画のエリアを分割することもできます。
# shareyはshare Y軸の意味です
fig, (axis1,axis2) = plt.subplots(1, 2, sharey=True)

sns.regplot(tips, x="total_bill", y="tip_pect", ax=axis1)
sns.violinplot(tips,
               x='size',
               y='tip_pect',
               ax=axis2)
```
