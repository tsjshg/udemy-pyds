```python
from numpy.random import randn
import seaborn as sns
```

```python
dataset = randn(100)

# ヒストグラムとカーネル密度推定を描きます
sns.histplot(dataset, bins=25, kde=True)
```

```python
# 確率密度にする
sns.histplot(dataset, bins=25, kde=True, stat='density')
```

```python
# rug plotとヒストグラム
sns.displot(dataset, rug=True)
```

```python
from pandas import Series

ser1 = Series(dataset, name='My_DATA')
```

```python
# pandasのSeriesを描画します
sns.histplot(ser1, bins=25, kde=True)
```
